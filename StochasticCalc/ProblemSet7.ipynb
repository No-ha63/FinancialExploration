```python
import numpy as np
import matplotlib.pyplot as plt
import math as m
from IPython.display import display, Math
```

## 4.2

```python
diff = 0.01
x = np.arange(0,1+diff,diff)

linespace = np.linspace(0,4,10000)

func = (2/m.sqrt(2* m.pi)) * np.exp(-1 * linespace**2 / 2)

sample = np.random.normal(loc = 0, scale = m.sqrt(diff), size= (10000,len(x) - 1))

sample = np.append(np.zeros(shape=(10000,1)), sample, axis= 1)

BM = np.cumsum(sample, axis = 1)

plt.plot(x,BM.T)
plt.title(r"$B_{t}$")
plt.xlabel("Time")
plt.ylabel("Value")
plt.show()

max = np.max(BM,axis = 1)

plt.hist(max, bins = 50, density=True)
plt.title(r"Histogram of $max_{s \leq 1}B_{s}$")
plt.ylim(0, 1.5)
plt.xlabel(r"Max value of $B_s$")
plt.ylabel("Density")
plt.show()

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

ax1.hist(max, bins = 50, density=True)
ax1.set_title(r"Histogram of $max_{s \leq 1}B_{s}$")
ax1.set_ylim(0, 1.5)
ax1.set_xlabel(r"Max value of $B_s$")
ax1.set_ylabel("Density")

ax2.hist(np.abs(BM[:, -1]), bins = 50, density=True)
ax2.set_title(r"Histogram of $|B_1|$")
ax2.set_ylim(0, 1.5)
ax2.set_xlabel(r"$|B_1|$ Value")
ax2.set_ylabel("Density")

plt.show()



plt.hist(np.abs(BM[:, -1]), bins = 50, density=True)
plt.title(r"Histogram of $|B_1|$")
plt.xlabel(r"$|B_1|$ Value")
plt.ylim(0, 1.5)
plt.ylabel("Density")
plt.show()


plt.hist(max, bins = 50, density=True, label = r'$max_{s \leq 1}B_{s}$', alpha = 0.5, color = 'b')
plt.hist(np.abs(BM[:, -1]), bins = 50, density=True, alpha = 0.6, label=r"$|B_1|$", color='y')
plt.title(r"Histogram of $|B_1|$ and $max_{s \leq 1}B_{s}$")
plt.ylim(0, 1.5)
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()

plt.hist(max, bins = 50, density=True, label=r"Historgam of $max_{s \leq 1}B_{s}$")
plt.plot(linespace, func, label = r"PDF of $|B_1|$")
plt.title(r"Histogram of $max_{s \leq 1}B_{s}$")
plt.xlabel(r"Max value of $B_s$")
plt.ylabel("Density")
plt.legend()
plt.show()
```

## 4.3

```python
diff = 0.01
x = np.arange(0,10+diff,diff)

sample = np.random.normal(loc = 0, scale = m.sqrt(diff), size= (10000,len(x) - 1))

sample = np.append(np.zeros(shape=(10000,1)), sample, axis= 1)

BM = np.cumsum(sample, axis = 1)

plt.plot(x,BM.T)
plt.title(r"$B_{t}$")
plt.xlabel("Time")
plt.ylabel("Value")
plt.show()

condition = BM >= 1
first_index = np.argmax(condition, axis=1)

first_index[first_index == 0] = -1

minT = x[first_index]


plt.hist(minT, bins = 45, density=True)
plt.title(r"Histogram of $\min(\tau,10)$")
plt.xlabel("Time")
plt.ylabel("Density")
plt.show()
```
