```python
import numpy as np
import matplotlib.pyplot as plt
import math as m
```

## 3.2

```python
diff = 0.01 
rate = 1
spacing = np.arange(0,10 + diff,diff)

samples = np.random.poisson(lam=(rate * diff), size=(10,len(spacing) - 1))
samples = np.append(np.zeros(shape=(10,1)), samples, axis=1)

final = np.cumsum(samples,axis=1)

plt.plot(spacing,final.T)
plt.title("Poisson Process for Rate 1 and Step 0.01")
plt.xlabel("Time")
plt.ylabel("Count")
#plt.grid(True)
plt.show()
```

## 3.3

```python
spacing = np.arange(0,1+diff,diff)
path_samples = np.random.normal(loc = 0, scale = m.sqrt(diff), size = (1000,len(spacing) - 1))
path_samples = np.append(np.zeros(shape=(1000,1)),path_samples, axis=1)

final = np.cumsum(path_samples,axis=1)

plt.plot(spacing,final.T, alpha = 0.3)
plt.show()
```

```python
# print(final)
# print(np.delete(final,0,1))
positive = final >= 0
counts = np.count_nonzero(positive,axis=1)
ratio = counts / (len(spacing))

spacing = np.linspace(0,1,1000)
funtion = (1/np.sqrt(spacing * (1 - spacing))) * (1/m.pi)
CFD = (2/m.pi) * np.arcsin(np.sqrt(spacing))

plt.hist(ratio,bins=51,density=True,label = "Ratio")
plt.plot(spacing,funtion,label = "Function")
plt.title("Histogram of Ratio of Time Spent Positive")
plt.xlabel("Ratio")
plt.ylabel("Density")
plt.legend()
plt.show()

plt.hist(ratio,bins=510,density=True,cumulative=True, label = "Ratio")
plt.plot(spacing,CFD,label = "CDF")
plt.title("CDF of Ratio of Time Spent Positive")
plt.xlabel("Ratio")
plt.ylabel("Density")
plt.legend()
plt.show()
```
