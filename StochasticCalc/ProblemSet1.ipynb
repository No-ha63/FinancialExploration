```python
N = 10000
samples = np.random.exponential(scale=1, size=N)
sums = np.cumsum(samples)
means = sums / np.arange(1, N + 1)

plt.plot(np.arange(1 , N + 1), means)
plt.axhline(y=1, color='orange', linewidth=0.5)
plt.show()
```

```python
def empirical_mean(N = 10000, runs = 10000):
    samples = np.random.exponential(scale = 1, size = (runs,N))
    return np.mean(samples, axis=1)

samples_100 = empirical_mean(100, 10000)
samples_10000 = empirical_mean(10000, 10000)


plt.hist(samples_100, bins=50, density= True)
plt.xlim(.5,1.5)
plt.title("100 PDF")
plt.show()

plt.hist(samples_100, bins=50, density= True, cumulative= True)
plt.xlim(.5,1.5)
plt.title("100 CDF")
plt.show()

plt.hist(samples_10000, bins=50, density= True)
plt.xlim(.5,1.5)
plt.title("10000 PDF")
plt.show()

plt.hist(samples_10000, bins=50, density= True, cumulative= True)
plt.xlim(.5,1.5)
plt.title("10000 CDF")
plt.show()
```

## 1.3

```python
def central_value(N = 100, size = 10000):
    lamb = 1
    std = m.sqrt(1/(lamb ** 2))
    samples = np.random.exponential(scale= 1/lamb, size=(size,N)) #the scale input is 1/lambda compared to how we define the exp dist
    sums = np.sum(samples, axis=1)
    return (sums - N * (1/lamb))/ (std * m.sqrt(N))

exponential = central_value()
normal = np.random.normal(loc=0, scale=1, size=10000)

fig,(ax1,ax2) = plt.subplots(nrows=1, ncols=2, figsize=(10,4))

ax1.hist(exponential, bins=50, density=True, color='blue')
ax1.set_title("Exponential")

ax2.hist(normal, bins=50, density=True, color='orange')
ax2.set_title("Normal")

plt.tight_layout()
plt.show()

plt.hist(exponential, bins=50, density=True, color='blue', alpha=0.5, label="Exponential")
plt.hist(normal, bins=50, density=True, color='yellow', alpha=0.5, label='Normal')
plt.legend(loc='upper right')
plt.title("Exponential vs Normal Samples")
plt.show()
```

## 1.4

```python
def cauchy(N = 10000):
    samples = np.random.rand(N)
    return np.tan(m.pi * (samples - (1/2)))

size = 10000
cauch = cauchy(size)

plt.hist(cauch, bins = 25000, density=True)
plt.xlim(-10,10)
plt.title("Cauchy Distribution Samples")
plt.show()
```

```python
fig,(ax1,ax2) = plt.subplots(nrows=1, ncols=2, figsize=(10,4))

ax1.hist(cauch, bins=25000, density=True, color='blue')
ax1.set_xlim(-10,10)
ax1.set_title("Cauchy")

ax2.hist(normal, bins=50, density=True, color='orange')
ax2.set_xlim(-10,10)
ax2.set_title("Normal")

plt.tight_layout()
plt.show()

plt.hist(cauch, bins=25000, density=True, color='blue', alpha=0.5, label="Cauchy")
plt.hist(normal, bins=50, density=True, color='yellow', alpha=0.5, label='Normal')
plt.xlim(-10,10)
plt.legend(loc='upper right')
plt.title("Cauchy vs Normal Samples")
plt.show()
```

```python
cumulative_sums = np.cumsum(cauch)
emp_means = cumulative_sums / np.arange(1, size + 1)

plt.plot(emp_means)
plt.title("Cauchy Empirical Means")
plt.show()
```

```python
def empirical_mean_cauchy(N = 10000, runs = 10000):
    samples = np.random.uniform(low=0, high=1, size = (runs,N))
    cauchy_samples = np.tan(m.pi * (samples - (1/2)))
    return np.mean(cauchy_samples, axis=1)

cauchy_10 = empirical_mean_cauchy(10, 10000)
cauchy_100 = empirical_mean_cauchy(100, 10000)


#plt.hist(cauchy_10, bins=5000, density= True)
plt.hist(cauchy_10, bins=50, density= True)
#plt.xlim(-10,10)
plt.title("Cauchy 10 PDF")
plt.show()

#plt.hist(cauchy_100, bins=10000, density= True)
#plt.xlim(-10,10)
plt.hist(cauchy_100, bins=50, density= True)
plt.title("Cauchy 100 PDF")
plt.show()
```
